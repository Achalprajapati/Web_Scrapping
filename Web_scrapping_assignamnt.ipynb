```python
Content = [] # Emptly list to storing the content later 

for i in range(1,10):# decideing the number of pages need to extract .
    url = f"http://quotes.toscrape.com/page/{i}/" # fetching the   htlm formatted data from web
    response = requests.get(url) # sending request to the website 
    response = response.content  # getting the response from website 
    soup = BeautifulSoup(response,'html.parser') # making the soup of the content for  bueautifyte data .
    webpage = soup.find('div') # getting the data from div class 
    contents = webpage.find_all('div',class_ = 'quote') # finding the content from  c
    for content in contents:
        author = content.find('small',class_ ="author").text
        quotes = content.find('span').text
        tag = content.find('meta',class_='keywords')
        tag=tag.attrs['content']
       
        Content.append([quotes,author,tag])
```

```python
df=pd.DataFrame(Content,columns=['quote','Name','Tag'])
```
